```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the CSV file
df = pd.read_csv('data/hh_income_district.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
print(df.columns)
df = df[df['date'] == '2022-01-01']
```

```python
# Create a histogram using Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(df['income_mean'], bins=30, color='skyblue', edgecolor='black')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.xticks(ticks=np.arange(0, 15001, 500), rotation=90)
plt.show()
```
